```python
import pandas as pd

student_data={
    "Roll No.":[101,102,103,104,105],
    "Telugu":[35,98,43,65,78],
    "English":[87,45,98,65,34],
    "Hindi":[45,26,78,45,67],
    "Maths":[45,56,67,78,89],
    "science":[23,34,45,56,76],
    "Social":[32,43,54,65,67]
}
data=pd.DataFrame(student_data)

Hall_Ticket_No=int(input("Enter the Roll No. : "))
index=data[data["Roll No."] == Hall_Ticket_No].index

score_card=data.loc[[index[0]]]
score_card.pop("Roll No.")
per=score_card.loc[index[[0]]].sum()*100/600

fail=score_card[score_card< 35].count()
no_of_fail_sub=fail.sum()
failed_sub=fail[fail==1].index

if(no_of_fail_sub>0):
  print("Status:Fail")
  print("Failed subjects:",failed_sub)

else:
  print("Status:Pass")
  print("Pass Percentage:",per)

score_card
```
